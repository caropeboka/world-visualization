```python
import folium
import pandas as pd
```

```python
data = pd.read_csv('drive/MyDrive/Dataset/world_country_and_usa_states_latitude_and_longitude_values.csv')
```

```python
#data pre processing
data.isna().sum()
```

```python
#drop null values
df = data.dropna(axis=0)
```

```python
df.isna().sum()
```

```python
place = df[ ['latitude', 'longitude']]
```

```python
type(place)
```

```python
place = place.values.tolist()
```

```python
type(place)
```

```python
map_country = folium.Map(location=[38, -98], zoom_start=4)
```

```python
def country(i,color):
  folium.Marker(
      location=point,
      popup=df['country'][i],
      tooltip='Click!',
      icon=folium.Icon(color=color, icon='tint', icon_color='white')
  ).add_to(map_country)
```

```python
i=0
for point in place:
  country(i,'red')
  i+=1
```
